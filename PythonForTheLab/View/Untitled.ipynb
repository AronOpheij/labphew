```python
import requests
from bs4 import BeautifulSoup
import re

from dateutil.parser import parse as date_parse

BASE_URL = 'https://arxiv.org'
```

```python
r = requests.get(BASE_URL + '/list/astro-ph/new')
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
papers = soup.find_all('span', {'class': 'list-identifier'})
```

```python
links = []
for paper in papers:
    link = paper.find('a', {'title': 'Abstract'})['href']
    links.append(link)
```

```python
paper_dates = []
for link in links:
    print('Processing: {}'.format(link))
    r = requests.get(BASE_URL + link)
    soup = BeautifulSoup(r.text, 'html.parser')
    history = soup.find('div', {'class': 'submission-history'})
    ans = history.find('b', text=re.compile('[v1]'))
    date_info = ans.nextSibling
    date_first_publish = date_info.partition('(')[0].partition(',')[2]
    paper_dates.append(date_parse(date_first_publish))
    print('Got date: {}'.format(date_first_publish))
```
